#DATA CLEANING:-

```python
#Dropping the columns which are not necessary or irrelevant to dataset.
df=df.drop(["url","phone","address","reviews_list","dish_liked","menu_item","listed_in(city)"],axis= 1)
```

```python
df
```

```python
#So finally checking shape after dropping the columns.
df.shape
```

```python
#Checking duplicates in the dataset.
df.duplicated().value_counts()
```

```python
#Dropping the duplicate rows from the dataset.
df.drop_duplicates(inplace=True)
```

```python
#So finally checking shape after dropping duplicate rows.
df.shape
```

```python
#Now,to see the null values in the columns.
df.info()
```

```python
#So, as we can see we have null values in rate,location,rest_type,cuisines,approx_cost(for two people).  
df.isna().sum()
```

```python
#showing null values in the columns in graphical form.
null_value=pd.DataFrame((df.isna().sum())*100/df.shape[0]).reset_index()
plt.figure(figsize=(16,7))
sns.pointplot('index',0,data=null_value)
plt.xticks(rotation=90,fontsize=15)
plt.title('percentage of null value')
plt.ylabel('percent')
plt.show()
```

##Lets clean the null values of each column one by one.

```python
#Checking the value counts of rate column.
df.rate.value_counts()
```

```python
#Checking unique values present in the rate column.
df.rate.unique()
```

```python
#As we can see there is 'NEW','nan','-' present in the column and also the denomenators, so we have to remove this.
def correct_rate(value):
  if(value=='NEW' or value=='-'):
    return np.nan
  else:
    value= str(value).split('/')
    value= value[0]
    return float(value)


df['rate']=df['rate'].apply(correct_rate)
df.rate.head()
```

```python
#Cleaning the null values by mean in the rate column.
df.rate.fillna(df['rate'].mean(),inplace=True)
```

```python
#Checking the value counts of location column.
df.location.value_counts()
```

```python
#Checking unique values present in the location column.
df.location.unique()
```

```python
#Cleaning the null values by 'Not present' in the location column.
df.location.fillna('Not Present',inplace=True)
```

```python
#Checking the value counts of rest_type column.
df.rest_type.value_counts()
```

```python
#Checking unique values present in the rest_type column.
df.rest_type.unique()
```

```python
#Cleaning the null values by 'Not present' in the rest_type column.
df.rest_type.fillna('Not present',inplace=True)
```

```python
#Checking the value counts of cuisines column.
df.cuisines.value_counts()
```

```python
#Checking unique values present in the cuisines column.
df.cuisines.unique()
```

```python
#Cleaning the null values by 'Not present' in the cuisines column.
df.cuisines.fillna('Not present',inplace=True)
```

```python
#Renaming the approx_cost(for two people to cost_per_two_people as it is unappropriate.
df.rename(columns={'approx_cost(for two people)':'cost_per_two_people'},inplace=True)
```

```python
#Checking the value counts of cost_per_two_people column.
df.cost_per_two_people.value_counts()
```

```python
#Checking unique values present in the cost_per_two_people column.
df.cost_per_two_people.unique()
```

```python
#As we can see there is commas in between the thousands value, so we need to remove that.
def remove_comma(value):
   value = str(value)
   if ',' in value:
     value = value.replace(',','')
     return float(value)
   else:
     return float(value) 


df['cost_per_two_people']=df['cost_per_two_people'].apply(remove_comma)
df.cost_per_two_people.head()
```

```python
#so the unique values of cost_per_two_people have changed to float and commas has been removed.
df.cost_per_two_people.unique()
```

```python
#Cleaning the null values by mean in the cost_per_two_people column.
df.cost_per_two_people.fillna(df['cost_per_two_people'].mean(),inplace=True)
```

*  Hence, all null values are cleaned from the columns. i.e rate,location,rest_type,cuisines,cost_per_two_people.
*  Data is all clean now.

```python
#Hence there is no null values present in the data set.
df.isna().sum()
```

```python
#Showing that their is no null values in the dataset using heatmap.
plt.figure(figsize=(14, 5))
sns.heatmap(df.isnull(), cbar=True, yticklabels=False)
plt.xlabel("column_name", size=14, weight="bold")
plt.title("missing values in column",fontweight="bold",size=17)
plt.show()
```

#**DATA VISUALISATION:-**

```python
#Checking how many online orders are placed and how many are not.
df.online_order.value_counts()
```

```python
#Checking top 10 restaurant types according to their ratings.
df_rating=df.groupby('rest_type').sum()['rate'].sort_values(ascending=False)[:10]
df_rating
```

```python
#Finding cost for two peoples by their restaurant types.
cost_by_rest_type=df.groupby('rest_type').sum()['cost_per_two_people'].sort_values(ascending=False)[:10]
cost_by_rest_type
```

```python
#Showing cost for two peoples by their restaurant types by using barh graph.
cost_by_rest_type.plot(kind="barh",figsize=(25,15),fontsize=15,color="m")
```
